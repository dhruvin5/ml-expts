```python
import pandas as pd
import numpy as np
data_frame = pd.read_csv('titanic_dataset.csv')

data_frame.head()
```

```python
data_frame.describe()
```

```python
positive_data =  data_frame.loc[(data_frame["Survived"] == 1)]
positive_data.describe()
```

```python
def findS(train_set,new_shuffled_data):
  hypothesis = train_set[0]
  print("The intial hypothesis is:", hypothesis)
  print("---------------------------------------------")
  for example in train_set:
    for i in range(len(example)):
      if hypothesis[i] != example[i] and hypothesis[i] != '?':
        hypothesis[i]='?'
  print("\nThe final hypothesis is:",hypothesis)
  print("---------------------------------------------")
  total_pass=0
  for example in new_shuffled_data:
    #print(example)
    for i in range(len(example)):
      if hypothesis[i] != example[i] and hypothesis[i] != '?':
        total_pass = total_pass + 1
  #print(len(shuffled_data))
  print("\nThe accuracy is: ",((len(shuffled_data)-total_pass)/len(shuffled_data))*100)
  print("---------------------------------------------")
```

```python
shuffled_data = positive_data.sample(frac=1)
shuffled_data.head()
new_data = shuffled_data[:4]
new_data.head()
```

```python
train_set = np.array(new_data)[:,2:] 
new_shuffled_data = np.array(shuffled_data)[:,2:] 
findS(train_set,new_shuffled_data)
```

Conclusion:
Since the FindS algorithm would give a generalized 
hypothesis for all the training examples,I took 4 random training examples from the dataset and ran FindS algorithm on it.

We can conclude that for a small training set FindS algorithm might give a more specific hypothesis.

Based on the above hypothesis we can conclude that people who embarked from 'S' have survived.
